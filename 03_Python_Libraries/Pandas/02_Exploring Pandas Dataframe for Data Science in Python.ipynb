# Pandas DataFrame-
- 2D size-mutable, potentially heterogenous tabular data structure with labeled axes(rows and columns).
- data in rows and columns.
- data, rows, and columns are the three primary components of a pandas dataframe.
- data rep is like a SQL table or a spreadsheet.

# Pandas DataFrame Constructor- 
Syntax : pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
- data: the dataframe data is specified using this arg. it might be a dictionary, list of dictionaries, a numpy array, or another dataframe, among other data structures. empty dataframe if data=None
- index: the row labels for the dataframe are specified using this argument. it could be a list, an array, or another dataframe. default is integer index if index is not specified
- columns: column labels are specified using this option. it could be a list, an array or another dataframe, among other data structures. a default set of column labels is created if column are not specified.
- dtype: the data type for te columns in the dataframe is specified using the dtype option. it is a dictionary that maps column names to data types or a single data type, such as int or float.
- copy: whether to copy the data , use the copy parameter. a new copy of the data is made if copy=True. original data used if copy=False.

## Parameters---

### 1. random_state - reproducible randomness
Concept:
A seed that controls the random number generator. Same seed = same random sample every time.

Use Case:
- Debugging: Get consistent results while developing

- Reproducible research: Others can replicate your results

- A/B testing: Ensure consistent group assignmen

Syntax: 
- df.sample(n=int_value, random_state=integer)  # Any integer works
- df.sample(n=int_value, random_state=None)     # Default - truly random

```python
# without random_state--> different results each time
import pandas as pd
import numpy as np

df=pd.read_csv('students.csv')

# WITHOUT random_state (different each run)
print("Without random_state (changes each time):")
print(df.sample(n=3))  # Run this multiple times - different results!
print(df.sample(n=3))  # Different from above

# WITH random_state (same each run)
print("\nWith random_state=42 (same every time):")
print(df.sample(n=3, random_state=42))
print(df.sample(n=3, random_state=42))  # Same as above!

# Any integer works as seed
print("\nDifferent seeds produce different results:")
print(df.sample(n=2, random_state=1))
print(df.sample(n=2, random_state=2))
print(df.sample(n=2, random_state=3))

# Common practice: Use a fixed seed for production
TRAIN_TEST_SEED = 2024
train_data = df.sample(frac=0.8, random_state=TRAIN_TEST_SEED)
test_data = df.drop(train_data.index)
```

### 2. replace - Sampling with/without Replacement

Syntax: df.sample(n=int_value, replace=True/False)

concept-
- replace=False ---- each row can be selected only once. like drawing random balls from a bag without replacing them.
- replace=True  ---- each row can be selected multiple times. like drawing names from a hat and putting them back each time.

```python
import pandas as pd
df_small=pd.DataFrame({'Item':['A','B','C','D']})
print("without replacement (no duplicates): ")
print(df_small.sample(n=3, replace=False, random_state=42))
print("-"*50)
```

Without replacement it cannot meet the demanded size if more than original

```python
df_small.sample(n=5,replace=False) # Error 
```

With Replacement

```python
print(df_small.sample(n=5, replace=True, random_state=42))
print("-"*50)
print()
# Real-world example: Bootstrap sampling
original_data = pd.DataFrame({'Value': range(10)})

# Create bootstrap sample (same size as original, with replacement)
bootstrap_sample = original_data.sample(n=len(original_data), replace=True, random_state=42)
print(f"\nBootstrap sample size: {len(bootstrap_sample)}")
print("May contain duplicates:", bootstrap_sample['Value'].duplicated().any())
```

Important concept- Bootstrapping.
It is sampling with replacement to estimate statistics

```python
import pandas as pd
import numpy as np
data=pd.DataFrame({'Measurement':np.random.normal(100,15,1000)})
# creating 1000 boorstrap samples to estimate confidence interval
bootstrap_means=[]
for i in range(1000):
    bootstrap_sample=data.sample(n=len(data),replace=True, random_state=i)
    bootstrap_means.append(bootstrap_sample['Measurement'].mean())
confidence_interval=np.percentile(bootstrap_means, [2.5,97.5])
print(f"95% Confidence Interval: {confidence_interval[0]:.2f}")
```

### 4. weights parameter (Probability Weights)

df.sample(n=int_value, weights='column_name'or [list of weights]) # Using Column as weights.

Concept-
- controls the probability of each row being selected. rows with higher weights have higher chance of being chosen.

Use Case-
- Stratified sampling: Ensure representation from different groups
- Importance sampling: Focus on more important records
- Biased sampling: Oversample rare events

```python
# eg 1 using a column as weights
# Students with higher scores should have higher chance of being selected
df_students = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Score': [50, 70, 90, 60, 85],
    'Grade': ['C', 'B', 'A', 'C', 'B']
})

# using score as weight (higher score = higher chance)
print("Sampling with score as weight (favors high scores): ")
print(df_students.sample(n=3, weights='Score', random_state=42))

# showing probabilities
total_score=df_students["Score"].sum()
df_students['Probability']=df_students['Score']/total_score
print(f"Selection probability for each student: \n{df_students}")
```

```python
# Give different importance to different groups
df_products = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Desk'],
    'Category': ['Electronics', 'Accessory', 'Accessory', 'Electronics', 'Furniture'],
    'Sales': [1000, 500, 300, 800, 200]
})

# Custom weights: Electronics 50% chance, Accessory 30%, Furniture 20%
def create_weights(category):
    weights_map = {'Electronics': 0.5, 'Accessory': 0.3, 'Furniture': 0.2}
    return weights_map[category]

df_products['CustomWeight'] = df_products['Category'].apply(create_weights)

print("Products with custom weights:")
print(df_products[['Product', 'Category', 'CustomWeight']])

print("\nWeighted sample (favors Electronics):")
print(df_products.sample(n=2, weights='CustomWeight', random_state=50))
```

### 6. axis(rows vs columns)
- Syntax:
  
  df.sample(n=int_value, axis=0) # Sample rows (default)
  
  df.sample(n=int_value, axis=1) # Sample columns

- Concept:

  axis=0 (default sample rows) (observations)

  axis=1: sample columns (features)

- Use:

  - feature selection: randommly select columns for testing
  - dimensionality reduction: get subset of features
  - random projection: create random feature combinations

```python
# e.g.
import pandas as pd
df_wide=pd.DataFrame({
    'A':range(5),
    'B': range(5, 10),
    'C': range(10, 15),
    'D': range(15, 20),
    'E': range(20, 25),
    'F': range(25, 30),
    'G': range(30, 35),
    'H': range(35, 40)
})

print("Original columns: ",list(df_wide.columns))
print("-"*50)

# sample 3 random columns
print(df_wide.sample(n=3,random_state=42, axis=1))
print("-"*50)

# sample 50% of columns 
print(df_wide.sample(frac=0.5, axis=1, random_state=42))
print("-"*50)

# sample columns with replacement (can get duplicates)
print(df_wide.sample(n=5, axis=1, replace=True, random_state=42))
print("-"*50)
```

### 7. ignore_index (Reset Index)
- Syntax:

  - df.sample(n=int_value, ignore_index=False)
  - df.sample(n=int_value, ignore_index=True)

- Concept:

  - ignore_index=False: keep the original index values from the source DataFrame
  - ignore_index-True: discard original indices and create new sequential indices (0,1,2.....)

- Use:

   - Creating new datasets: when you want clean 0-based indexing
   - Mergine samples: To avoid index conflicts
   - Exporting data: for cleaner output files

```python
# e.g. 
import pandas as pd
df_custom= pd.DataFrame({
    'Product':['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Desk'],
    'Price':[1000, 25, 75, 300, 450]
},index=['P001', 'P002', 'P003', 'P004', 'P005'])

print("Original DataFrame with custom indices:")
print(df_custom)
print("-"*50)

# without ignore index
print(df_custom.sample(n=3, random_state=42))
print("-"*50)

# with ignore_index (resets to 0,1,2)
print(df_custom.sample(n=3, ignore_index=True, random_state=42))
print("-"*50)

# real world: 
```

method used- dropna()
- using this method, there are more number of options provided due to multiple columns
- rows from df removed where atleast 1 missing value.

```python
# Syntax: 
DataFrame.dropna(
    axis=0,              # 0=rows, 1=columns
    how='any',           # 'any' or 'all'
    thresh=None,         # minimum number of non-NA values
    subset=None,         # specific columns/rows to check
    inplace=False,       # modify original DataFrame?
    ignore_index=False   # reset index after dropping?
)
```

# parameters - 

```python
# 1. axis- drop rows or columns
import pandas as pd
import numpy as np

# Create sample data with missing values
df=pd.DataFrame({
    'A':[1, 2, np.nan, 4],
    'B':[5, np.nan, 7, 8],
    'C':[9, 10, 11, np.nan]
})

print(f"Original DataFrame:")
print(df)
print("_"*50)
print()

# axis=0 (default)- drop rows
# drop rows with any missing values
df_rows=df.dropna(axis=0)
print("Drop Rows (axis=0)")
print(df_rows)
print("_"*50)
print()

# axis=1 Drop Columns:
# drop columns with any missing value
df_cols=df.dropna(axis=1)
print("Drop columns(axis=1)")
print(df_cols)
# No column remains.
# axis='rows' and axis='columns' can also be used instead of 0 and 1 for readability
print("="*50)
print()
```

```python
# 2. how - 'any' vs 'all'

df1 = pd.DataFrame({
    'A': [1, 2, np.nan, np.nan],
    'B': [5, np.nan, np.nan, np.nan],
    'C': [9, 10, 11, np.nan]
})
print(f"df1: \n{df1}")
print("_"*50)
# how='any' (by default): drop if any value missing

# drop rows where any column has NaN
result_any=df1.dropna(how='any')
print(result_any)
# only row 0 remains
print("_"*50)
print()

# how='all' drops only if all values are missing
# drop rows where all columns are NaN
result_all=df1.dropna(how="all")
print(result_all)
print("="*50)

# Row 0,1,2 remain (row 3 had all NaN, so it's dropped)
# Survey data where some participants skipped entire sections
survey = pd.DataFrame({
    'Q1': [1, 2, np.nan, np.nan],
    'Q2': [3, 4, np.nan, np.nan],
    'Q3': [5, 6, np.nan, np.nan],
    'Comments': ['Good', 'OK', np.nan, 'Nice']
})

# Drop only completely empty responses (all answers missing)
cleaned = survey.dropna(how='all')
print("Keep partial responses:")
print(cleaned)
# Row 2 (all NaN except Comments? Actually Comments also NaN) dropped
# But row 0,1,3 kept
```

```python
# 1. working with different types of missing values
# pandas recognizes multiple types of missing values
df=pd.DataFrame({
    'A': [1, 2, None, 4],        # None
    'B': [5, 6, np.nan, 8],      # np.nan
    'C': [9, 10, pd.NA, 12],     # pd.NA
    'D': ['x', 'y', '', 'z']     # Empty string (NOT missing!)
})
print(f"df: \n{df}")
print("-"*50)
print(df.isna())
# Note: empty string '' is not detexted as missing
# use: df['D']=df['D'].replace('',pd.NA) to convert 
```

```python
# 2. Chaining dropna() with other methods
# cleaning data in one pipeline
df=pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Price': [100, np.nan, 300, np.nan, 500],
    'Quantity': [10, 20, np.nan, 40, 50],
    'Date': ['2024-01-01', '2024-01-02', np.nan, '2024-01-04', '2024-01-05']
})
print("="*50)
# Chain operations
cleaned = (df
    .dropna(subset=['Price'])           # Remove rows without price
    .dropna(thresh=2)                   # Keep rows with at least 2 values
    .reset_index(drop=True)             # Reset index
)
print(f"cleaned: \n{cleaned}")
```

```python
# 3. Dropping vs Filling - Decision Guide
# When to drop vs when to fill
# Scenario 1: Critical column has missing data -> DROP
customer_data = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Email': ['a@x.com', np.nan, 'c@x.com', 'd@x.com'],
    'Purchase': [100, 200, 300, 400]
})

# Email is critical for marketing - drop missing
clean_customers= customer_data.dropna(subset=['Email'])

# Scenario 2: Non-Critical column, few missing -> FILL
sales_data = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 200, np.nan, 400],
    'Notes': ['Good', 'OK', np.nan, 'Great']
})
# Only 1 missing sales value- fill with mean 
sales_data['Sales'].fillna(sales_data['Sales'].mean(), inplace=True)

# Scenario 3 : Too many missing values -> drop column
df = pd.DataFrame({
    'Important': [1, 2, 3, 4],
    'MostlyMissing': [np.nan, np.nan, 3, np.nan],  # 75% missing
    'Complete': [5, 6, 7, 8]
})
# Drop column with >50% missing
missing_percentage=df['MostlyMissing'].isna().mean()
if missing_percentage > 0.5 : 
    df.drop('MostlyMissing',axis=1, inplace=True)
```

```python
# Concept 4: Using dropna() with Time Series
# Time series often requires special handling
dates=pd.date_range('2024-01-01',periods=10)
ts_data=pd.Series([1, 2, np.nan, np.nan, 5, 6, np.nan, 8, 9, 10], index=dates)
print(f"ts_data: \n{ts_data}")
print("-"*50)

# Dont just drop- you lose time continuity
ts_dropped=ts_data.dropna()
print(ts_dropped)

# Better: Interpolate or use forward fill
ts_filled=ts_data.interpolate()
print(ts_filled)
```

## major pitfalls

```python
# PITFALL 1: Empty strings aren't NaN
df = pd.DataFrame({'A': ['', 'value', np.nan]})
print("Empty string not detected as NaN:")
print(df.isna())  # Only row 2 is True

# SOLUTION: Convert empty strings to NaN
df['A'] = df['A'].replace('', pd.NA)
df.dropna(inplace=True)

# PITFALL 2: Forgetting that dropna returns a new DataFrame by default
df = pd.DataFrame({'A': [1, np.nan, 3]})
df.dropna()  # This does nothing to df
print(df)  # Still has NaN!

# SOLUTION: Assign or use inplace
df = df.dropna()  # Assign
# OR
df.dropna(inplace=True)  # Inplace

# PITFALL 3: Using thresh incorrectly
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, np.nan, 6], 'C': [7, 8, np.nan]})
# thresh=3 means keep rows with ≥3 non-NA (all columns must be non-NA)
result = df.dropna(thresh=3)
print("thresh=3 (same as how='any'):")
print(result)  # Only row 0

# thresh=2 means keep rows with at least 2 non-NA
result = df.dropna(thresh=2)
print("\nthresh=2 (keep if at least 2 values present):")
print(result)  # All rows kept

# PITFALL 4: Not handling mixed missing value types
df = pd.DataFrame({
    'A': [1, None, 3],      # None
    'B': [4, np.nan, 6],    # np.nan
    'C': ['x', pd.NA, 'z']  # pd.NA
})

# This works fine - all detected
df.dropna(inplace=True)
```

# Arithmetic Operators for DataFrames :-
core concept - 
- artithmetic cannot be performed directly on a DataFrame because it has multiple columns, but you can on series (individual columns).

```python
# Basic operations
import pandas as pd

df = pd.DataFrame({
    'Employee Name': ['John', 'Jane', 'Bob'],
    'Salary': [50000, 60000, 75000]
})

print("Original DataFrame: ")
print(df)
print("_"*80)

# Method 1:
# adding 5000 to everyone's salary
df['Salary']+5000

# Subtract 2000
df['Salary']-2000

# Multiply by 1.1 (10% raise)
df['Salary']*1.1

#Divide by 12 (monthly salary)
df['Salary']/12


# Method 2: using methods - .add(), .sub(), .mul(), .div()
# Same operations but with methods
df['Salary'].add(5000)      # Addition
df['Salary'].sub(2000)      # Subtraction
df['Salary'].mul(1.1)       # Multiplication
df['Salary'].div(12)        # Division
```

```python
# Important addition concepts not in book
print("Method 1")
# 1. Operations between 2 columns
df=pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob'],
    'Salary': [50000, 60000, 75000],
    'Bonus': [5000, 6000, 7500]
})
print(f"original df : \n{df}")
print("-"*50)
print()

# Add two columns
df['Total Compensation']=df['Salary']+df['Bonus']

# Calculate percentage
df['Bonus Percentage']=(df['Bonus']/df['Salary'])*100
print(df)
print("="*50)

print("Method 2")
# 2. Operations with Conditions
## Give 10% raise only to those earning < 60000
df['New Salary']=df['Salary'].apply(lambda x: x*1.1 if x< 60000 else x)

## or using numpy
import numpy as np
df['Adjusted Salary']=np.where(df['Salary']<60000,
                              df['Salary']*1.1,
                              df['Salary'])
print(df)
print("="*50)

print("Method 3")
# 3. Arithmetic with Entire DataFrame
## you can do arithmetic on entire DataFrame if all columns are numeric
df_numeric=pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Add 10 to every value in DataFrame
print(df_numeric+10)
print("_"*50)

# Multiplying entire dataframe by 2
print(df_numeric*2)
print("_"*50)
# Operations between two DataFrames
df2=pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60],
    'C': [70, 80, 90]
})
print(df_numeric+df2)
print("="*50)

print("Method 4")
# 4. using radd() and other reversed operations
# where order matters (for subtraction and division)
df=pd.DataFrame({'A':[10,20,30]})

# regular subtraction: A- 5
print(df['A'].sub(5))
print("_"*50)

# reversed subtraction : 5-A
print(df['A'].rsub(5))
print("_"*50)

# Regular division: A/2
print(df['A'].div(2))
print("_"*50)

# Reversed division: 2/A
print(df['A'].rdiv(2))
print("="*50)

print("Method 5")
# Handling missing values in Arithmetic
df = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob'],
    'Salary': [50000, np.nan, 75000]
})

# Result will have NaN where original had NaN
print(df['Salary'] + 5000)
print("_"*80)

# Fill NaN before arithmetic
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print(df['Salary'] + 5000)  # Now all values are valid
print("_"*80)
```

#### 2. Stable Sorting-

preserves original order for equal values (dosent shuffle ties randomly)

```python
df=pd.DataFrame({
    'Group': ['A', 'A', 'A', 'B', 'B'],
    'Original_Pos': [1, 2, 3, 4, 5],
    'Value': [30, 10, 20, 50, 40]
})
print(f"df: \n{df}")
print("_"*50)

# stable sort ( preserves 1,2,3 order for Group A)
df_stable=df.sort_values(by='Group', kind='stable')
print(f"stale_sorted_df: \n{df_stable}")
print("-"*50)

# Unstable sort (might shuffle A's randomly)
df_unstable=df.sort_values(by='Group', kind='quicksort')
print(f"quick_sorted_df: \n{df_unstable}")

# Use when: You need predictable order for equal values (time series, sequential data, reproducible results)
```

#### 3. Sorting with numpy.argsort() for complex logic

returns indices that would sort the array- allows multi stage custom sorting

```python
import numpy as np
df=pd.DataFrame({'Salary': [50000, 150000, 75000, 200000, 30000]})
print(f"df: \n{df}")
print("_"*50)

# Complex logic: High earners (>100k) first, then by salary within each group
is_high_earner = (df['Salary'] > 100000).astype(int)  # 1 for high, 0 for low
print(f"is_high_earner: \n{is_high_earner}")
print("_"*50)

# argsort returns index positions
# Negative = descending order
order = np.argsort([-is_high_earner, -df['Salary']], axis=0)[::-1]
print(f"order: \n{order}")
print("_"*50)

df_sorted = df.iloc[order[0]]
print(f"df_sorted: {df_sorted}")


# how it works in simple words
# 1. argsort() returns indices: [2,0,1] means "put index 2 first then 0 then 1"
# 2. use iloc[] to reorder DataFrame
# 3. Multiple conditions= list of arrays
# Use when complex conditional logic needed that sort_Values cant solve
```

#### 4. Sorting with nlargest() and nsmallest()-

gets top/bottom N values without sorting the entire dataset.

```python
# Fast: gets top 3 directly (dosent sort all rows)
top3=df.nlargest(3, 'Salary')
print(f"top3: {top3}")
print("_"*50)

# Slow : Sorts everything then takes first 3
top3_slow = df.sort_values ('Salary', ascending=False).head(3)
print(top3_slow)
print("_"*50)

# Get bottom 2
bottom2=df.nsmallest(2,'Salary')
print(f"bottom 2: {bottom2}")
print("_"*50)

# Per Group (top 2 per department)
top_per_dept= df.groupby('Department').apply(lambda x: x.nlargest(2, 'Salary'))
print(top_per_dept)
print("_"*50)

# Performance Difference:
# - nlargest()= scans once
# - sort_values().head()= sorts everything
# - 100x faster for large datasets!
# Use when: You only need top/bottom N, not full sorted data
```

```python
# e.g.(s)
import pandas as pd
import numpy as np

df =pd.DataFrame({
    'Student':['Alice','Bob','Charlie','David','Emma'],
    'Score':[95,85,95,90,85]
})
print(f"df: \n{df}")
print("-"*50)

# method 1: 'average' (default)
df['rank_avg']=df['Score'].rank(method='average',ascending=False)
print(df[['Student','Score','rank_avg']])
#   Student  Score  rank_avg
# 0   Alice     95       1.5  # (1+2)/2 = 1.5  -
# 1     Bob     85       4.5  # (4+5)/2 = 4.5  |
# 2 Charlie     95       1.5  # (1+2)/2 = 1.5  |-- how it is assigned
# 3   David     90       3.0  # Only one 90    |
# 4    Emma     85       4.5  # (4+5)/2 = 4.5  -
print("-"*50)

# method 2: 'min'
df['rank_min']=df['Score'].rank(method='min',ascending=False)
print(df[['Student','Score','rank_min']])
print("-"*50)

# method 3: 'max'
df['rank_max']=df['Score'].rank(method='max',ascending=False)
print(df[['Student','Score','rank_max']])
print("-"*50)

# method 4: 'first' it dosent skip or duplicate ranks, but assigns unique ranks based on the order the data appears
df['rank_first']=df['Score'].rank(method='first', ascending=False)
print(df[['Student','Score','rank_first']])
# here no gaps are seen and no duplicates come... ties broken by who appeared first to the reader
print("-"*50)

# method 5: 'dense' dosent skip ranks and handles ties cleanly.
df['rank_dense'] = df['Score'].rank(method='dense', ascending=False)
print(df[['Student', 'Score', 'rank_dense']])
print("-"*50)
```

```python
# important parameters

# 1. ascending
# ascending = True - lowest value gets rank 1
df['rank_low_to_high']=df['Score'].rank(ascending=True)

# ascending = False - highest value gets rank 1
df['rank_high_to_low']=df['Score'].rank(ascending=False)

# 2. na_option
# handling missing values
df_with_nan= pd.DataFrame({'Score': [95, np.nan, 85, 90, np.nan]})

# keep (default): NaN stays NaN
df_with_nan['rank_keep']=df_with_nan['Score'].rank(na_option='keep')

# top: NaN gets highest rank (rank 1)
df_with_nan['rank_top']=df_with_nan['Score'].rank(na_option='top')

# bottom: NaN gets lowest rank (last rank possible)
df_with_nan['rank_bottom']=df_with_nan['Score'].rank(na_option='bottom')

# 3. percentile rank
df['percentile']=df['Score'].rank(pct=True)

print(df_with_nan)
```

```python
# Technique 1: Filtering with multiple conditions (Right Way)
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Salary': [75000, 52000, 90000, 65000, 48000],
    'Experience': [3, 5, 7, 2, 4]
})

# multiple conditions - Always use paranthesis ()
result=df[
    (df['Department'].isin(['IT','HR'])) & # DEPARTMENT IT or HR
    (df['Salary'] > 50000) &
    (df['Experience'] >= 3)
]
print(result)
print("="*50)
```

### Common Filtering Mistakes

```python
# 1. Forgetting Parantheses
# ❌ WRONG - This causes error
df[df['A'] > 5 & df['B'] < 10]

# ✅ CORRECT
df[(df['A'] > 5) & (df['B'] < 10)]

# 2. Using = instead of ==
# ❌ WRONG - Assignment, not comparison
df[df['City'] = 'Durg']

# ✅ CORRECT
df[df['City'] == 'Durg']

# 3. Chained Indexing (The Silent Bug)
# ❌ BAD - May fail silently or give warning
df[df['Salary'] > 50000]['Bonus'] = 1000

# ✅ GOOD - Use .loc
df.loc[df['Salary'] > 50000, 'Bonus'] = 1000

# 4. Case Sensitivity
# ❌ 'DURG' won't match 'Durg'
df[df['City'] == 'DURG']

# ✅ Convert to same case
df[df['City'].str.lower() == 'durg']

# Or use case=False in contains
df[df['City'].str.contains('durg', case=False)]

# 5. Using isin() with wrong type
# ❌ Comparing int column with string list
df[df['Salary'].isin(['50000', '60000'])]  # Won't match!

# ✅ Same type
df[df['Salary'].isin([50000, 60000])]
```

### Optimizations (tips)

```python
# 1. Filter Before Selecting Columns
# ❌ SLOW - Selects all columns first
df[df['Salary'] > 50000][['Name', 'Salary']]

# ✅ FAST - Filter first, then select columns
df.loc[df['Salary'] > 50000, ['Name', 'Salary']]
#---------------------------------------------------------------------------------------------------------

# 2. Use query() for Large Data
# query() is optimized with numexpr
result = df.query('Salary > 50000 and Department == "IT"')
#---------------------------------------------------------------------------------------------------------

# 3. Avoid Chained Conditions When Possible
# ❌ Multiple filters (creates intermediate copies)
temp = df[df['Salary'] > 50000]
result = temp[temp['Department'] == 'IT']

# ✅ Single filter
result = df[(df['Salary'] > 50000) & (df['Department'] == 'IT')]
#---------------------------------------------------------------------------------------------------------

# 4. Use isin() for Multiple OR Conditions
# ❌ SLOW - Many OR conditions
df[(df['City'] == 'A') | (df['City'] == 'B') | (df['City'] == 'C')]

# ✅ FAST - One isin() call
df[df['City'].isin(['A', 'B', 'C'])]
#---------------------------------------------------------------------------------------------------------
```
