# Array Creation using linspace() function

linspace(start,stop,num=50,endpoint=True,retstep=False,dtype=None)
start:the number with which it starts(optional), default=0 
Stop:end point of interval rannge.
num:indicates the number of samples to generate.
endpoint:if set to False, the stop value is excluded and when rue, it is included.
retstep: spacing or steps between the value.
dtype: specifies the output array type.

```python
import numpy as mynp
print(mynp.linspace(0,2))
print('-------------'*5)
print(mynp.linspace(0,2,3))
print('-------------'*5)
print(mynp.linspace(0,5,5,endpoint=False))
print('-------------'*5)
print(mynp.linspace(1,5,5,endpoint=True))
print('-------------'*5)
print(mynp.linspace(0,5,5,endpoint=False,retstep=True))
print('-------------'*5)
print(mynp.linspace(0,5,5,endpoint=False,retstep=True,dtype=int))
```

# Utilizing diag() function

# Some additional functions of random()
## 1.rand()- generates uniformly distributed 'float' values in range(0,1)

## 2.uniform()-generates uniformly distributed 'float' values in range(low,high)

## 3.randn()- generates normally distributed float values with a mean of zero and a standard deviation of one.
### What this looks like in practice:
When you call randn() repeatedly, you'll get values that are mostly clustered around 0, with these probabilities:
About 68% of values will be between -1 and +1
About 95% of values will be between -2 and +2
About 99.7% of values will be between -3 and +3

### It's commonly used for:
Adding noise to data
Initializing neural network weights
Statistical simulations
Testing algorithms with realistic random variations

## 4.normal()- generates normally distributed float values with a specified mean and a stamndard deviation.
The normal() function generates random numbers from a normal (Gaussian) distribution where you specify both the mean and standard deviation yourself.
### Syntax:
normal(mean, standard_deviation, size=None)
### Key difference from randn():
randn() = fixed mean=0, std=1
normal() = you choose the mean and std
### Common use cases:
- Simulating real-world data: heights, weights, test scores
- Adding specific noise: sensor readings with known error margins
- Financial modeling: returns with specific volatility
- Scientific simulations: measurements with known uncertainty

# Elements access of ndarray
elements of ndarray can be accessed using 
- indexing
- slicing
- condition based selection
- advanced indexing

```python
arr=np.random.randint(1,100,size=(3,3,3))  # a 3D array
print(arr)
print("-----------"*5)
# 1. Using indexing: single element access, zero based indexing supporting both +ve and -ve
print(f"arr[0][0][1]={arr[0][0][1]}")
print(f"arr[2][2][2]={arr[2][2][2]}")
print("-----------"*5)
# 2. Using slicing: 
print("last element of all matrices =")
print(f"{arr[:,2:,2:]}")
print("-----------"*5)
```

# 3. Using advanced indexing: 
- Integer indexing
- Boolean indexing
## Integer indexing
 ----- Syntax: arr[[2-D array indices],[row indices],[column indices]]
 its like a(i,j) we learnt in matrix-the ( element index ).this is true for 2D arrays.
 for 3D array just one more index comes, i.e. the 2D array index before (i,j)-- i.e. (2D-array-index, i, j) 

```python
arr3d=np.arange(27).reshape(3,3,3)
print(arr3d)
print("-------"*5)
print(arr3d[([0,1],[1,2],[1,2])])
# this array returned contains elements: 
# [(0 position 3D array, 1st row x 1st column),
# (1 position 3D array, 2nd row x 2nd column)]
```

## Boolean indexing

```python
arr3d=np.arange(27).reshape(3,3,3)
print(arr3d)
print("-------"*5)
print(arr3d>5)              # A bollean array of true and false for fuklfilling conditions
print("-------"*5)
print(arr3d[arr3d>5])        # gives the 1D-array of elements fulfilling the condition
print("-------"*5)
# To retain the shape of the array and not let it convert to 1D array use where method
print(np.where(arr3d>5,arr3d,0))  # 1st value- condition, 2nd value- what to return if condition is true, 3rd value- what to return if condition false
print("-------"*5)
# To apply multiple conditions to array use: logical_and
print(np.logical_and(arr3d>5,arr3d<24))
print("-------"*5)
# to print the array with multiple conditions:
conditions=np.logical_and(arr3d>5,arr3d<24)
print(arr3d[conditions])    # returns a 1D-array
```

# Iterate elements of ndarray
- Iteration using Python loop
- Iteration using nditer() function:
- Iteration using ndenumerate() function

## 1.Iteration using Python loop

```python
# 3D-Array
import numpy as np
arr=np.arange(27).reshape(3,3,3)
print(arr)
print("-------"*5)
for outer in arr:  # iterating 2D arrey
    for inner in outer:  # iterating 1D array
        for data in inner:  # Iterating through scalar values in 1D array
            print(data)
print("-------"*5)
```

## 2. Iteration using nditer() function:

```python
import numpy as np
arr=np.arange(1,7).reshape(2,3)
print(arr)
print("-------"*5)
for loop in np.nditer(arr):
    print(loop)
print("-------"*5)
```

## 3.Iteration using ndenumerate() function

```python
import numpy as np
# this funtion returns indexes in addition to elements as multidimensional
for myposition,myvalue in np.ndenumerate(arr):
    print(f"{myvalue} is on position:{myposition}")
    
```

# Arithmetic operators in numpy

```python
import numpy as np
arr1=np.random.randint(1,10,size=(3,3))
arr2=np.random.randint(11,20,size=(3,3))
print(f"arr1 :\n{arr1}")
print(f"arr2 :\n{arr2}")
print("-----------"*5)
# size
print(f"size arr1={arr1.size}")
print(f"size arr2= {arr2.size}")
# shape
print(f"shape arr1= {arr1.shape}")
print(f"shape arr2= {arr2.shape}")
# dimension
print(f"dimension arr1= {arr1.ndim}")
print(f"dimension arr2= {arr2.ndim}")
```

```python
import numpy as np
# Addition
print(f"Addition of arr1,arr2=\n {np.add(arr1,arr2)}")
# Subtraction
print(f"Subtraction of arr1,arr2=\n {np.subtract(arr1,arr2)}")
# Multiplication
print(f"Multiplication of arr1,arr2=\n {np.multiply(arr1,arr2)}")
# Division
print(f"Division of arr1,arr2=\n {np.divide(arr1,arr2)}")
# Floor Division
print(f"Floor Division of arr1,arr2=\n {np.floor_divide(arr1,arr2)}")
# Modulus
print(f"Modulus of arr1,arr2=\n {np.mod(arr1,arr2)}")
# Power
print(f"arr1 power arr2=\n {np.power(arr1,arr2)}")
```
