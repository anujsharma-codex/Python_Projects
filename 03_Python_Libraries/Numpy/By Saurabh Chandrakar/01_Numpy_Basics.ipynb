# Broadcasting concept in numpy
- arithmetic ops are typically carried out between 2 arrays that share the same dimension, shape, size.
- some mathematic ops are allowed even with some variation in shape size and dimension
- broadcasting is automatically done by numpy.
- not every situation lends to broadcasting.
- numpy has principles for doing broadcasting. it happens only when those rules are met.
  this is important for while dealing with arrays of varying dimensions, shapes or sizes

## Rules
### Rule 1
ensure dinensions are same.
When two arrays have different numbers of dimensions, NumPy adds dimensions of size 1 to the beginning (left side) of the smaller-dimensional array until both arrays have the same number of dimensions.

## Rule 2
When two arrays have the same number of dimensions (after Rule 1), but the sizes in specific dimensions don't match, NumPy stretches any dimension of size 1 to match the corresponding dimension in the other array.


```python
#example 
import numpy as np
# After Rule 1, we have:
mynp1 = np.array([[1, 2, 3],      # shape: (2, 3)
                  [4, 5, 6]])

mynp2 = np.array([[10, 20, 30]])  # shape: (1, 3) - after Rule 1

print(f"mynp1 shape: {mynp1.shape}")  # (2, 3)
print(f"mynp2 shape: {mynp2.shape}")  # (1, 3)

# Now apply Rule 2 - the dimension of size 1 gets "stretched"
result = mynp1 + mynp2

# Visual Step-by-Step Process

Step 1: Before Stretching
mynp1 (2,3):        mynp2 (1,3):
[ [1, 2, 3]         [ [10, 20, 30] ]   ← Only 1 row
  [4, 5, 6] ]

Step 2: During Stretching (Conceptual)
# mynp2 gets "replicated" along the dimension of size 1:

Original mynp2:     After conceptual stretching:
[ [10, 20, 30] ] →  [ [10, 20, 30],
                      [10, 20, 30] ]    ← Row replicated

Step 3: After Stretching (The Operation)
mynp1:              mynp2 (stretched conceptually):
[1, 2, 3]     +     [10, 20, 30]   =   [11, 22, 33]
[4, 5, 6]           [10, 20, 30]       [14, 25, 36]

#example 2- column stretching:
arr1 = np.array([[1, 2, 3],         # shape: (2, 3)
                 [4, 5, 6]])

arr2 = np.array([[10],               # shape: (2, 1) - after Rule 1
                 [20]])

# arr2 gets stretched horizontally (columns)
# Result: arr2 becomes conceptually (2, 3)
result = arr1 + arr2
# [[11, 12, 13],    # 1+10, 2+10, 3+10
#  [24, 25, 26]]    # 4+20, 5+20, 6+20

#example 3- Both Arrays Have Size 1 Dimensions
arr1 = np.array([[1, 2, 3]])        # shape: (1, 3)
arr2 = np.array([[10],               # shape: (2, 1)
                 [20]])

# Both get stretched!
# arr1 stretches vertically: (1,3) → (2,3)
# arr2 stretches horizontally: (2,1) → (2,3)
result = arr1 + arr2
# [[11, 12, 13],    # 1+10, 2+10, 3+10
#  [21, 22, 23]]    # 1+20, 2+20, 3+20
```

Note: If the sizes in any dimension are not the same and are not equal
to one, we will receive an error because numpy is unable to do
broadcasting across those arrays.

```python
#Incompatible Example
arr1 = np.ones((3, 2))     # shape: (3, 2)
arr2 = np.ones((3, 3))     # shape: (3, 3)

# After Rule 1 (already same dimensions)
# Compare rightmost: 2 vs 3 - NOT equal, and neither is 1
# result = arr1 + arr2  # This will raise ValueError!
```

IN RULE 1 HOW TO DECIDE WHICH ARRAY HAS LESS DIMENSION AND IN RULE 2 WHICH ARRAY GETS EXPANDED OUT OF THE 2, LIKE IF BOTH TRY TO REPLICATE EACH OTHER, NO ONE WILL BE ABLE TO REPLICATE EACH OTHER.  ???

```python
#Rule 1: Identifying the Array with Fewer Dimensions
#NumPy automatically determines this by comparing the ndim (number of dimensions) of each array.

import numpy as np

arr1 = np.ones((2, 3, 4))     # ndim = 3
arr2 = np.ones((5, 6))         # ndim = 2
arr3 = np.ones(7)              # ndim = 1
scalar = 5                     # ndim = 0

print(f"arr1 dimensions: {arr1.ndim}")  # 3
print(f"arr2 dimensions: {arr2.ndim}")  # 2
print(f"arr3 dimensions: {arr3.ndim}")  # 1
print(f"scalar dimensions: {np.array(scalar).ndim}")  # 0

# Rule 1 is applied automatically:
# The array with FEWER dimensions gets 1's padded on the LEFT
```

```python
#Rule 2: Which Array Gets Expanded?
#Critical clarification: BOTH arrays can be expanded, but only in dimensions where they have size 1!

#Scenario 1: Only One Array Has Size 1 :-
arr1 = np.ones((2, 3))        # shape: (2, 3)
arr2 = np.ones((1, 3))        # shape: (1, 3) - after Rule 1

# arr2 gets expanded (its first dimension is 1)
# arr1 stays the same (no dimension of size 1)
result = arr1 + arr2

#Scenario 2: BOTH Arrays Have Size 1 in Different Dimensions
arr1 = np.ones((2, 1))        # shape: (2, 1) - 1 in column dimension
arr2 = np.ones((1, 3))        # shape: (1, 3) - 1 in row dimension

# BOTH arrays get expanded!
# arr1 expands horizontally: (2,1) → (2,3)
# arr2 expands vertically: (1,3) → (2,3)
result = arr1 + arr2
print(result.shape)  # (2, 3)

#Visual of BOTH expanding:
arr1 (2,1):     arr2 (1,3):     Result (2,3):
[1]             [10, 20, 30]    After BOTH expand:
[2]                             [1+10, 1+20, 1+30] = [11, 21, 31]
                                [2+10, 2+20, 2+30] = [12, 22, 32]

arr1 expands:   arr2 expands:
[1, 1, 1]       [10, 20, 30]
[2, 2, 2]       [10, 20, 30]
```

##### The Golden Rule of Broadcasting
For each dimension (comparing right to left), at least ONE of these must be true:

The dimensions are equal

One dimension is 1 (that array will stretch)

One array doesn't have that dimension (effectively size 1 after Rule 1)

##### Complete Example Showing Both Rules

```python
# Step-by-step broadcasting
arr1 = np.ones((4, 3))        # shape: (4, 3)
arr2 = np.ones(3)              # shape: (3,)

print(f"Before Rule 1: arr1 {arr1.shape}, arr2 {arr2.shape}")

# RULE 1: arr2 has fewer dimensions (1 vs 2)
# Pad arr2 on LEFT with 1's: (3,) → (1, 3)
# After Rule 1 (conceptual): arr1 (4,3), arr2 (1,3)

# RULE 2: Compare dimensions from right to left
# Rightmost: 3 vs 3 → equal ✓
# Next: 4 vs 1 → one is 1 (arr2) → arr2 expands
# arr2 expands vertically: (1,3) → (4,3)

result = arr1 + arr2  # Works!
print(f"Result shape: {result.shape}")  # (4, 3)
```

1. Adding Bias in Neural Networks

```python
# Features: 100 samples, 3 features
X = np.random.randn(100, 3)     # shape: (100, 3)

# Bias term: 1 value per feature
bias = np.array([0.1, 0.2, 0.3])  # shape: (3,)

# Add bias to EVERY sample (100 times!)
X_with_bias = X + bias  
# bias gets stretched: (3,) → (1,3) → (100,3)

# Without broadcasting, you'd need:
# X_with_bias = X + bias.reshape(1,3).repeat(100, axis=0)
```

2. Centering Data (Subtracting Mean)

```python
# Dataset: 1000 samples, 50 features
data = np.random.randn(1000, 50)     # shape: (1000, 50)

# Calculate mean of each feature (50 means)
feature_means = data.mean(axis=0)     # shape: (50,)

# Center the data - subtract means from ALL samples
centered_data = data - feature_means  
# feature_means stretches: (50,) → (1,50) → (1000,50)

# This is EXACTLY how data preprocessing works!
```

3. Normalization/Standardization

```python
# Data: 1000 samples, 50 features
data = np.random.randn(1000, 50)

# Calculate statistics
mean = data.mean(axis=0)      # shape: (50,)
std = data.std(axis=0)         # shape: (50,)

# Standardize: (data - mean) / std
# ALL THREE arrays broadcast together!
normalized = (data - mean) / std  
# mean: (50,) → (1,50) → (1000,50)
# std:  (50,) → (1,50) → (1000,50)
```

4. Scaling Images (Computer Vision)

```python
# Batch of RGB images: 32 images, 64x64 pixels, 3 channels
images = np.random.randn(32, 64, 64, 3)  # shape: (32, 64, 64, 3)

# Per-channel scaling factors (e.g., ImageNet normalization)
channel_means = np.array([0.485, 0.456, 0.406])  # shape: (3,)
channel_stds = np.array([0.229, 0.224, 0.225])   # shape: (3,)

# Apply to EVERY pixel in EVERY image
normalized_images = (images - channel_means) / channel_stds  
# channel_means: (3,) → (1,1,1,3) → (32,64,64,3)
```

7. Weighted Average Example

```python
# Student grades: 30 students, 5 assignments
grades = np.random.randint(50, 100, size=(30, 5))  # shape: (30, 5)

# Assignment weights (different importance)
weights = np.array([0.1, 0.2, 0.3, 0.25, 0.15])    # shape: (5,)

# Weighted grades for ALL students
weighted_grades = grades * weights  
# weights: (5,) → (1,5) → (30,5)
final_scores = weighted_grades.sum(axis=1)
```

##### Why This Pattern Is So Common
In data science, we constantly have:

- Main data: Many samples/observations (large dimension)
- Parameters/statistics: One set of values for all observations (dimension with size 1 that stretches)
