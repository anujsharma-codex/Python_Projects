# PANDAS SERIES-
- its a 1D labeled array 
- stores any sort of data (Python objects, floats, strings, or integers)
- Pandas series equivalent to column in the Excel Sheet.
- Series objects- mutable.
- series size- immutable.

## PANDAS SERIES CONSTRUCTOR

- data:
  information kept in series. scalar value, dictionary, list, or array numpy.
  
- index:
  this details the series index labels. could be dictionary, list, array, or range. series is given a default integer index
  (starting 0 if no index given).

- dtype:
  defines data type of values. pandas tries to determines on its own if not given as input.

- name:
  series name is specified by this argument.

- copy:
  user can decide wheather the input data should be copied.

- fastpath:
  most users should not use this argument because it is used for internal optimization only.

## Accessing series values using head and tail methods

head()
- accesses the fist five rows of a series as the default value is five. if argument says n then it will access first n rows.
  Syntax: head(n:'int'=5)

tail()
- accesses last 5 rows as default value. if n is provided, it accesses last n rowsof a series.
  if n value: -ve, function returns all rows except first n rows.

```python
import pandas as pd
ser=pd.Series([n for n in range(20)])
print(f"ser: \n {ser}")
print("-"*50)
print(f"head method n=none: \n{ser.head()}")
print("-"*50)
print(f"head method n=2: \n{ser.head(n=2)}")
print("-"*50)
print(f"tail method n=none: \n{ser.tail()}")
print("-"*50)
print(f"tail method n=-18: \n{ser.tail(n=-18)}")  # equivalent to below
print("-"*50)
print(f"tail method n=2: \n{ser.tail(n=2)}")
print("-"*50)
```

# Accessing elements in Pandas Series-
- individual elements in a series can be accessed using either an index label or the element's integer location.
- Syntax: ser[x]
- x can be index value.ser[x]. then it returns the value present at that index.
- x can be list of indices.ser[[1,2,3]]. then is returns a series of values present at indices in the list...1,2,3.
- x can be slices as well. ser[2:4]. returns values from 2nd index to 4.

```python
import pandas as pd
lst=[11,12,13,14,15]
ser=pd.Series(lst,index=['r','s','t','u','v'])
print(f"ser: \n{ser}")
print("-"*50)

# using index labels
print("ser['u'] : ")
print(ser['u'])
print("-"*50)
# element access using integer location
print("ser['3'] : ")
print(ser[3])
```

## Values extraction using loc and iloc indexers :-
- for extracting values using labels based selection,--- use of loc indexer.
- for extracting on basis of position based selection, iloc is used. iloc= (integer loc)
- using these nhances performances.

```python
import pandas as pd
lst=[11,12,13,14,15]
ser=pd.Series(lst,index=['r','s','t','u','v'])
print(f"ser: \n{ser}")

# using iloc
print(f"ser.iloc[1]:  \n{ser.iloc[1]}")
print("-"*50)
print(f"ser.iloc[[1,3]]:  \n{ser.iloc[[1,3]]}")
print("-"*50)
print(f"ser.iloc[1:3]:  \n{ser.iloc[1:3]}")
print("-"*50)
print(f"ser.iloc[-1]:  \n{ser.iloc[-1]}")  # we are able to access the last item
print("="*50)

# using loc
print(f"ser.loc['s']:  \n{ser.loc['s']}")
print("-"*50)
print(f"ser.loc[['s','u']]:  \n{ser.loc[['s','u']]}")
print("-"*50)
print(f"ser.loc['s':'u']:  \n{ser.loc['s':'u']}")
print("-"*50)
```

```python
import pandas as pd
def my_even_number_selection(myseries):
    # Use myseries.size, not series
    return [True if loop % 2 == 0 else False for loop in range(myseries.size)]
series = pd.Series([loop for loop in range(10)])

print("Approach-1--------------------------")
print(series[my_even_number_selection(series)])  # ← Need to CALL the function

print("Approach-2--------------------------")
# This works - lambda returns boolean list
print(series.loc[lambda series_: [True if loop % 2 == 0 else False for loop in range(series_.size)]])

print("Approach-3--------------------------")
# Convert boolean to integer positions
indices = [i for i in range(series.size) if i % 2 == 0]
print(series.iloc[indices])
# OR use lambda that returns integer positions
print(series.iloc[lambda x: [i for i in range(x.size) if i % 2 == 0]])

print("Approach-4--------------------------")
# .get() doesn't work this way - removing it
# Just use direct boolean indexing
print(series[series % 2 == 0])  # Simplest way!
```

# Usage of the apply method to Pandas series :-
- can be applied to each element of a pandas series using the apply method.
- it can be a normal python function or a numpy function.
- basically it applies a predefined function on values in a pandas series.
- Syntax : Series.apply(func, convert_dtype=True, args=(),**kwargs)
- where....
  func: The function to apply (can be built-in, lambda, or custom)
  args: Tuple of positional arguments to pass to the function
  **kwargs: Keyword arguments to pass to the function

```python
import pandas as pd
lst=[11,12,13,14,15]
ser1=pd.Series(lst,index=['r','s','t','u','v'])

def cube(num):
    return num**3
ser1_cube=ser1.apply(cube)
print(ser1_cube)
print(ser1)
# print(help(ser1.apply))
```

# Series object transformation

## 1. Partial transformation

- in this ops ae performed on a subset of series / dataframe.
- it can be performed using update() method.
- Non NaN values will be used from the passed series to make updates.

```python
import pandas as pd

# wihout update method and instead using normal indexing or loc/iloc indexers
ser=pd.Series([15,30,45,60])
ser[1]=25
ser[2]=55
print(ser) # Obs: We can see that we are using 'n' lines of code, in order to update 'n' elements
print("-"*50)
print()
# Transforming series using update method
ser2=pd.Series([15,30,45,60])
ser2.update(pd.Series([25,55],index=[1,2]))
print(ser2)
```

- if the matched index are there then old values are replaced by new values.
- if matched index not there, and there are more indexes, then extra elements will be ignored.
- the corresponding values are not updated in the original series if the argument series contains NaNs.

```python
import pandas as pd
import numpy as np
orig_ser=pd.Series(['A','B','C','D','E'],
                  index=[1,2,3,4,5])
print(f"Original : \n{orig_ser}")
print("-"*50)
print()

# scenario 1 matched index - old values replaced
# done above

# scenario 2 no match and extra indices- Ignored
prompt=pd.Series(['X','Y','Z'],
                      index=[10,20,30])
orig_ser.update(prompt)
print(orig_ser)  #  NOTHING CHANGED because no indices matched
print("-"*50)
print()

# scenario 3: mixed - matching and extra indexes

mixed_prompt= pd.Series(['X', 'Y', 'Z', 'W'], 
                index=[2, 4, 6, 8])  # 2 and 4 match, 6 and 8 are extra
orig_ser.update(mixed_prompt)

print("\nAfter update:")
print(orig_ser)
# Extra indices (6 and 8) were IGNORED - not added to original
print("-"*50)
print()

# scenario 3 NaN Values - Not Updated
# if updating series has NaN values then the corresponding indexes in it are not updated and remain same
new_prompt=pd.Series(['X',pd.NA,'Z',pd.NA],
                    index=[1,2,4,5])
orig_ser.update(new_prompt)
print(orig_ser)
```

## Alternative: Using combine_first() for different behavior :-
- combines 2 series or dataframes
- more powerful. combines 2 Series/DataFrames, using values from the first object using values from first object, but filling missing values from the second.
- Syntax: Series.combine_first(other)

```python
s1=pd.Series([1,2,np.nan,4,np.nan],
             index=['a','b','c','d','e'])
print(f"s1 : \n{s1}")
print("-"*50)

s2=pd.Series([100,200,300],
             index=['c','e','f'])
print(f"s2: \n{s2}")
print("-"*50)

# combining s1 first then s2 for missing values
print(f"{s1.combine_first(s2)}")
# a      1.0
# b      2.0
# c    100.0  ← Filled from s2 (was NaN)
# d      4.0
# e    200.0  ← Filled from s2 (was NaN)
# f    300.0  ← NEW index added from s2!
# dtype: float64

#Key Difference from fillna(): combine_first() added index 'f' which wasn't in s1!
```

## Using fillna()
- filles missing values
- (NaN, NAN None) in a single series or dataframe.
- Syntax: Series.fillna(value=None, method=None, axis=None,
- inplace=False, limit=None)

```python
import pandas as pd
import numpy as np

# with a scalar value
s=pd.Series([1,2,np.nan,4,np.nan,6])
print(f"Original: \n{s}")
print("-"*50)
print(f" filled: {s.fillna(0)}")
print("-"*50)

# filling different values with different position
s = pd.Series([1, 2, np.nan, 4, np.nan, 6])

# Use of dictionary to specify different fill values per index
print(f"filled : \n{s.fillna({2:99, 4:100})}")
print("-"*50)
```

## Global Transformation :-
- operations get performed on entire series and dataframe.
- this operation can be performed on the entire series/dataframe.
- it can be done either using apply() or map()
- apply() already done before

## map() :-
- it is a limited version of apply()
- here extra arguments to the input function cannot be passed.
- here each value in a series will be substituted with another value which may be derived from a function, a dict, or a series.
- Syntax: map(arg:'Callable | Mapping | Series', na_action: "Literal['ignore'] | None"= None)

```python
import pandas as pd
import numpy as np

ser1= pd.Series(['Green','Yellow','Blue','Violet'])
print(f"ser1: \n{ser1}")
print("-"*50)
print()

# using dict
ser2=ser1.map({'Green':'Pink','Blue':'Magenta'})
print(F"ser2: \n{ser2}") # there is no replacement for Yellow and Violet color hence converted to NaN
print("-"*50)
print()

# Using lambda function
func=ser1.map(lambda x: f"I love {x} Colour")
print(f"func: {func}")
print("-"*50)
print()

# using na_action parameter
def func_2(x):
    return f"I LOVE {x} COLOUR"
ser3=pd.Series(['Green','Yellow',np.nan,'Violet'])
func2=ser3.map(func_2,na_action='ignore') # for keeping as NaN, na_action parameter is used so as to avoid applying the function to missing values
print(f"func2: \n{func2}")
print("-"*50)
print()

# so while partial transformation target a specific subset of the data,
# global transformation are applied to the entire data structure.
# Pandas offer a wide array of built-in functions and methods to efficiently
# perform both types of transformations.
```
